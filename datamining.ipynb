```python
# Mengimpor library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
 
# Mengimpor dataset
dataset = pd.read_csv('Data-Jumlah-Jemaah-Haji-yang-Berangkat-dan-Meninggal-Tahun-1997-2011.csv')
X = dataset.iloc[:, [2, 3]].values
print(X)
 
# Menjalankan K-Means Clustering ke dataset
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 0)
y_kmeans = kmeans.fit_predict(X)
dataset['cluster'] = y_kmeans
print(dataset)

# Visualisasi hasil clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 120, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 150, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 200, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 250, c = 'cyan', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='+', s = 250, c = 'yellow', label = 'Centroids')
plt.title('Clusters Jumlah Jemaah haji')
plt.xlabel('jumlah_korban_meninggal_di_arab_saudi')
plt.ylabel('jumlah_korban_meninggal_di_pesawat')
plt.legend()
plt.show()
```
